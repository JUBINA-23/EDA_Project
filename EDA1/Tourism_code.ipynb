# Title - Tourism Resource Allocation Management Dataset
### This dataset contains data on tourism management, with a focus on resource allocation and service delivery at various tourist destinations. The dataset includes information on visitor counts, environmental factors, resource type and availability, visitor reviews and other key indicators that affect tourist experiences. The purpose of this dataset is to support the optimization of resource allocation, improve service delivery, and enhance the tourist experience.

#### Format Data

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])                  # dtype - datetime
df['visitor_count'] = np.uint16(df['visitor_count'])               # dtype - uint16, visitor_count always +ve
df['avg_stay_duration'] = np.uint8(df['avg_stay_duration'])        # dtype - uint8, time in minutes always +ve
df['air_quality'] = np.uint8(df['air_quality'])                    # dtype - uint8, AQI always +ve
df['overcrowded'] = df['overcrowded'].astype(bool)                 # dtype - bool (T/F) 
df['rating'] = np.uint8(df['rating'])                              # dtype - uint8, rating 1 - 5 
```

#### Removing wrong data

```python
sns.boxplot(data=df)           # finding outliers using boxplot from seaborn 
plt.xticks(rotation = 90)
plt.show()
```

#### Finding duplicates

```python
print(df.duplicated())                        # search duplicates
print(df.drop_duplicates(inplace = True))     # remove duplicates
# no duplicates
```

### Description
#### The dataset contains 15 columns and 1,000 rows of data, with each row representing a specific timestamp and the corresponding conditions at a tourist location
##### 1     timestamp	         - The date and time when the data was collected
##### 2	    location_id          - unique id for the tourist destination or location
##### 3	    visitor_count	     - number of visitors at the destination at the given timestamp
##### 4	    avg_stay_duration    - average duration (in minutes) tourists spend at the location
##### 5	    peak_visitors	     - binary indicator representing whether the visitor count exceeded a threshold (0 or 1)
##### 6	    temperature	         - temperature at the location in Celsius
##### 7	    humidity	         - relative humidity percentage at the location
##### 8	    air_quality	         - air quality index (AQI) at the location, indicating pollution levels
##### 9	    weather_condition	 - current weather condition (e.g., Sunny, Cloudy, Rainy)
##### 10	resource_type	     - type of resource being tracked (e.g., restroom, food stall, parking spot)
##### 11	available_resources  - number of resources available at the location (e.g., available restrooms or parking spaces)
##### 12	resource_capacity    - total(max) capacity of resources available at the location
##### 13	resource_demand	     - current demand for a specific resource
##### 14	overcrowded          - True indicates overcrowding and False indicates non-overcrowded situation, based on visitor count and available resource
##### 15	rating               - location review given by visitors rated 1 - 5

### 1. What are the highest and lowest visit count of each location and on which dates?

```python
sorted_df = df.sort_values(by=['location_id','visitor_count'])
highest = sorted_df.drop_duplicates(subset = 'location_id',keep='last')[['location_id','visitor_count','timestamp']]
print('Highest visit count of each location\n\n',highest)
lowest = sorted_df.drop_duplicates(subset='location_id',keep='first')[['location_id','visitor_count','timestamp']]
print('\n\nLowest visit count of each location\n\n',lowest)

# barplot 
plt.bar(highest['location_id'],highest['visitor_count'], width = 0.2)    # highest
plt.xticks(df['location_id'])                                            # to display location_id 1,2,3 instead of 0.5,1.0,1.5,etc
plt.bar(lowest['location_id'],lowest['visitor_count'])                   # lowest
plt.title("Highest and lowest visit count of each locations")            # title
plt.xlabel("LOCATION ID")                                                # xlabel 
plt.ylabel("VISIT COUNT")                                                # ylabel
plt.show()                                                               # display graph
```

### 2. What is the total and average visitor count of each location?

```python
grouped = df.groupby('location_id')
total = grouped['visitor_count'].agg(np.sum)
average = grouped['visitor_count'].agg(np.mean)
print("Total visit count:\n",total)
print('\nAverage visit count:\n',average)
```

### 3. On which dates are most resources available in each location?

```python
sorted = df.sort_values(by = 'available_resources',ascending = False)
d1 = sorted.drop_duplicates(subset = 'location_id',keep = 'first')[['timestamp','location_id','available_resources']]
print(d1)

# piechart
plt.pie(d1['location_id'], labels = d1['available_resources'], startangle = 30,autopct ="%1.1f%%")
plt.title("Most available resources in each location\n")
plt.legend(d1['location_id'])
plt.show()
```

### 4. What is the maximum and minimum resource capacity in each location and what are the resource types there? 

```python
pivot = pd.pivot_table(
    data=df,
    index= 'location_id',
    values= ['resource_capacity','resource_type'],
    aggfunc=['max','min'])
print(pivot)
```

### 5. Which locations have the highest and lowest environment factors(temperature, humidity, air quality)?

```python
sort_temp = df.sort_values(by = ['location_id','temperature'])
high_temp = sort_temp.drop_duplicates(subset = 'location_id',keep = 'last')[['location_id','temperature']]
low_temp = sort_temp.drop_duplicates(subset = 'location_id',keep = 'first')[['location_id','temperature']]
print("Highest temperature\n",high_temp,"\nLowest temperature:\n",low_temp)


sort_hum = df.sort_values(by = ['location_id','humidity'])
high_hum = sort_hum.drop_duplicates(subset = 'location_id',keep = 'last')[['location_id','humidity']]
low_hum = sort_hum.drop_duplicates(subset = 'location_id',keep = 'first')[['location_id','humidity']]
print("\nHighest humidity:\n",high_hum,"\nLowest humidity:\n",low_hum)

sort_air = df.sort_values(by = ['location_id','air_quality'])
high_air = sort_air.drop_duplicates(subset = 'location_id',keep = 'last')[['location_id','air_quality']]
low_air = sort_air.drop_duplicates(subset = 'location_id',keep = 'first')[['location_id','air_quality']]
print("\nHighest AQI:\n",high_air,"\nLowest AQI:\n",low_air)
```

### 12. What is the total visit count of each resource type based on ratings?

```python
reso = pd.pivot_table(data = df, index = ['resource_type','rating'],values = ['visitor_count'], aggfunc = 'sum')
print(reso)

# line plot
reso_reset = reso.reset_index()                 # resetting the index to make resource_type a column
sns.lineplot(data= reso_reset, x='rating', y ='visitor_count', hue='resource_type',style ='resource_type', marker = 'o')
plt.title('Total Visit count of each Resource type by Rating\n')
plt.xlabel('\nRating')
plt.xticks(reso_reset['rating'])
plt.ylabel('Total Visitors\n')
plt.legend()
plt.show()
```

### 13. What is the count of rating 5 on each dates?

```python
# method 1
filtered = df.groupby('timestamp').filter(lambda x: len(x[x['rating'] == 5]) > 0)
filtered['timestamp'] = filtered['timestamp'].dt.date
rating_5_counts = filtered.groupby('timestamp')['rating'].apply(lambda x: (x == 5).sum()).reset_index(name='rating_5_count')
print(rating_5_counts)

# method 2
rate5= df[df['rating'] == 5]
rate5['timestamp'] = rate5['timestamp'].dt.date
rating_5 = rate5.groupby('timestamp').size().reset_index(name='rating_5_count')
print(rating_5)

# bar plot
sns.barplot(x='timestamp', y='rating_5_count', data = rating_5, palette = 'husl')
plt.xticks(rotation = 45)
plt.title("Count of Rating 5 on Each Dates\n")
plt.xlabel("Dates")
plt.ylabel("No. of Rating_5")
plt.show()
```

### 14. What is the total no. of visitor count on each day?

### 15. What is the average stay duration of each location?

### 16. On what factors does the stay duration depends?

### 17. Calculate the no. of days, each locations were overcrowded and not?

```python
crowd = df.groupby('location_id')['overcrowded'].agg(
    overcrowded ='sum',
    not_overcrowded = lambda x: (~x).sum()
)
print(crowd)

# double bar graph
crowd = crowd.reset_index()
crowd.set_index('location_id')[['overcrowded', 'not_overcrowded']].plot(kind='bar', width=0.8, color=['r', 'k'])
plt.title('No. of Overcrowded and Not Overcrowded Days by Location\n')
plt.xlabel('\nLocations')
plt.ylabel('Number of Days\n')
plt.xticks(rotation = 0)
plt.show()
```

### 18. What is the total and average no. of available resources of each resource type in each location?

```python
availres = pd.pivot_table(data = df, index = ['location_id','resource_type'],values = ['available_resources'], aggfunc = ['sum','mean'])
availres.columns = ['Total', 'Average']
print(availres)

# bar plot
# total
colors = ['r','b','g']
availres['Total'].plot(kind='bar',color = colors)
plt.title('Total Available Resources by Location and Resource Type\n')
plt.xlabel('\nLocations with resource type')
plt.ylabel('Total Available Resources\n')
plt.xticks(rotation=45)
plt.show()

# average
colors = ['pink','skyblue','lavender']
availres['Average'].plot(kind='bar', color = colors)
plt.title('Average Available Resources by Location and Resource Type\n')
plt.xlabel('\nLocations with resource type')
plt.ylabel('Average Available Resources')
plt.xticks(rotation=45)
plt.show()
```

### 19. What is the distribution of resource demand as a percentage of resource capacity for each location?

```python
res_dem = df.groupby("location_id")['resource_demand'].sum()
res_cap = df.groupby("location_id")['resource_capacity'].sum()
percent = (res_dem / res_cap) * 100
result = pd.DataFrame({
    'Resource Capacity': res_cap,
    'Resource Demand': res_dem,
    'Percent': percent
})
print(result)

# double bar plot
result[['Resource Capacity', 'Resource Demand']].plot(kind='bar', color = ['m','y'])
plt.xlabel('Locations')
plt.ylabel('Resource Values')
plt.title('Resource Demand vs Resource Capacity by Location')
plt.xticks(rotation = 0)
plt.legend( loc='upper right')
plt.show()
```

### 20. What percentage of resource capacity is being utilized by each resource type?


```python
res_dem = df.groupby("resource_type")['resource_demand'].sum()
res_cap = df.groupby("resource_type")['resource_capacity'].sum()
percent = (res_dem / res_cap) * 100
result = pd.DataFrame({
    'Resource Demand': res_dem,
    'Resource Capacity': res_cap,
    'Percent': percent
})
print(result)

# pie chart
result[['Percent']].plot(kind='bar', color = 'c')
plt.xlabel('\nResource Types')
plt.ylabel('Percent\n')
plt.title('Utilization percentage of each Resource type\n')
plt.xticks(rotation = 0)
plt.show()
```

### 21. What is the total peak visit count of each location, order by highest visit count with dates?

```python
peak1_df = df[df['peak_visitors'] == 1]             # consider only peak_visitors = 1 and not 0
peak = pd.pivot_table(data = peak1_df, index = ['location_id','timestamp'], values = 'visitor_count', aggfunc = 'sum')
sorted_peak = peak.sort_values(by=['location_id','visitor_count'], ascending=[True,False])
print(sorted_peak)

# bar plot
colors = ['m']
sorted_peak.plot(kind='bar', color = colors)
plt.title('Peak visit count by location and dates\n')
plt.xlabel('\nLocations with dates')
plt.ylabel('Total visit count')
plt.xticks(rotation=90)
plt.show()
```
