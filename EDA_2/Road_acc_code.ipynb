# Title - Road Accident Report Dataset
### This dataset contains data on road accidents, including location, severity, weather and road conditions. It also records the number of casualties, vehicles involved, and other key factors like speed limits and light conditions, etc. The purpose of this dataset is to analyze road accident patterns, identify contributing factors and improve traffic safety. It can support planning to reduce accidents, enhance road conditions and safety improvements in road traffic.

### Cleaning Data 
#### Rename Attribute

```python
df.rename(columns={'Accident Date' : 'Accident_Date', 'Local_Authority_(District)' : 'Local_Authority', 'Urban_or_Rural_Area' : 'Urban_or_Rural'},inplace = True)
```

#### Format Datatype

```python
df['Accident_Date'] = pd.to_datetime(df['Accident_Date'], format= '%d-%m-%Y')        # dtype - datetime
df['Time'] = pd.to_datetime(df['Time'], format= '%H:%M')                             # dtype - time
df['Number_of_Casualties'] = np.uint16(df['Number_of_Casualties'])                   # dtype - uint16, casualty count - always +ve
df['Number_of_Vehicles'] = np.uint8(df['Number_of_Vehicles'])                        # dtype - uint8, vehicle count - always +ve
```

#### Check Empty Cells

```python
print(df.isna().sum())
print(df.dropna(inplace = True))
```

#### Duplicates

```python
print(df.duplicated())                        # search duplicates
print(df.drop_duplicates(inplace = True))     # remove duplicates
# no duplicates
```

#### Wrong Data

```python
sns.boxplot(data=df)           # finding outliers using boxplot from seaborn 
plt.xticks(rotation = 90)
plt.show()
```

### Description
#### The dataset is related to road accidents and contains 3,00,494 entries with 19 columns as detailed below:
##### 1. Accident_Index        - A unique identifier for each accident.
##### 2. Accident_Date         - The date of the accident.
##### 3. Day_of_Week           - The day of the week when the accident occurred.
##### 4. Junction_Control      - The type of control at the junction, such as traffic signals or uncontrolled.
##### 5. Junction_Detail       - Details about the junction (e.g., T-junction, crossroads).
##### 6. Accident_Severity     - The severity of the accident (e.g., Serious, Slight).
##### 7. Latitude              - The latitude coordinate of the accident location.
##### 8. Light_Conditions      - The light conditions during the accident (e.g., daylight, darkness).
##### 9. Local_Authority       - The administrative district where the accident occurred.
##### 10. Longitude            - The longitude coordinate of the accident location.
##### 11. Number_of_Casualties - Number of casualties in the accident.
##### 12. Number_of_Vehicles   - Number of vehicles involved in the accident.
##### 13. Police_Force            - The police force handling the accident.
##### 14. Road_Surface_Conditions - The condition of the road surface (e.g., dry, wet).
##### 15. Road_Type               - The type of road where the accident occurred (e.g., single carriageway).
##### 16. Speed_limit             - Speed limit on the road where the accident occurred.
##### 17. Time                    - Time of the accident.
##### 18. Urban_or_Rural          - Whether the area is urban or rural.
##### 19. Weather_Conditions      - Weather conditions during the accident (e.g., fine, rainy).
##### 20. Vehicle_Type            - The type of vehicles involved (e.g., car, motorcycle)

### 1. Compare the count of accidents in the years 2021 and 2022?

```python
acc_count = (df['Accident_Date'].dt.year.value_counts().rename_axis("Year"))
print(acc_count)

# bar plot
acc_count.plot(kind = 'bar')
plt.title("Accident count in 2021 and 2022\n")
plt.xlabel("\nYear")
plt.ylabel("Accident count\n")
plt.xticks(rotation = 0)
```

### 2. Which day of the week have most accident count?

```python
week_count = df["Day_of_Week"].value_counts().reindex(['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
print(week_count)

# bar plot
colors = ['r','b','k','g','y','c','orange']
week_count.plot(kind = 'bar',color = colors)
plt.title("Total Accident Count on Each Days of the week\n")
plt.xlabel("\nDays of week")
plt.ylabel("Accident Count")
plt.show()
```

### 3. Categorize accident counts based on junction control and junction type.

```python
jn_count1 = pd.pivot_table(data = df, index=['Junction_Control'], values='Accident_Index', aggfunc='count').sort_values("Accident_Index")
jn_count1.columns = ['Accident Count']
print(jn_count1)

jn_count2 = pd.pivot_table(data = df, index=['Junction_Detail'], values='Accident_Index', aggfunc='count').sort_values("Accident_Index")
jn_count2.columns = ['Accident Count']
print(jn_count2)

# pie chart
explodes = [0.1,0.5,0.1,0.1,0.1,0.1]
plt.pie(jn_count1['Accident Count'], labels = jn_count1.index, autopct = "%1.1f%%", explode = explodes)
plt.title('\nAccident Count by Junction Control\n')
plt.show()

explodes = [0.5,0.5,0.5,0.5,0.5,0.2,0.2,0.1,0]
plt.pie(jn_count2['Accident Count'], labels = jn_count2.index, autopct = "%1.1f%%", explode = explodes)
plt.title('\nAccident Count by Junction Type\n\n\n')
plt.show()
```

### 4. What is the accident count based on severity?

```python
sev_count = df['Accident_Severity'].value_counts().rename_axis("Severity")
print(sev_count)

# pie chart
plt.pie(sev_count,labels = sev_count.index, autopct = '%1.1f%%')
plt.title("\nAccident count based on Severity")
plt.legend(sev_count)
plt.show()
```

### 5. What are the factors causing increase in casualty count?

```python
count_rel = df.corr(numeric_only = True)
print(count_rel['Number_of_Casualties'])

# bar plot
rela1 = count_rel['Number_of_Casualties'].drop('Number_of_Casualties')   # remove self
rela1.plot(kind='bar', color = ['g','r','g','g'])
plt.xticks(rotation = 90)
plt.title("\nFactors Affecting Casuality Count\n")
plt.xlabel("\n\nFactors")
plt.ylabel("Correlation\n")
plt.show()
```

### 6. What is the most and least number of casualty by location?

```python
cas_count = df.groupby("Local_Authority")['Number_of_Casualties'].agg([sum,max,min]).sort_values('sum')
print(cas_count)                   # to display total sum, max and min no. of casualty of each location
print(cas_count.iloc[0])           # display min
print(cas_count.iloc[-1])          # display max
```

### 7. What is the total accident count in each local authority?

### 8. How do accident count change with light conditions?

### 9. Which road surface conditions and road type cause more accidents?

```python
road_surf = df['Road_Surface_Conditions'].value_counts().sort_values(ascending = False)
print(road_surf)

road_type = df['Road_Type'].value_counts().sort_values(ascending = False)
print(road_type)

# pie chart
plt.pie(road_surf, labels = road_surf.index, autopct = '%1.1f%%')
plt.title("\nAccident Count distribution by Road surface conditions\n")
plt.legend(road_surf)
plt.show()

plt.pie(road_type, labels = road_type.index, autopct = '%1.1f%%')
plt.title("\nAccident Count distribution by Road type\n")
plt.legend(road_type)
plt.show()
```

### 10. Which area (Urban or rural) have more number of accidents?

```python
area_count = df['Urban_or_Rural'].value_counts().sort_values(ascending = False)
print(area_count)

# pie chart 
plt.pie(area_count, labels = area_count.index, autopct = "%1.1f%%")
plt.title("\nNo. of Accidents in Urban/Rural Areas\n")
plt.legend(area_count)
plt.show()
```

### 11. What is the total number of causalities and vehicles in the year 2021 and 2022?

```python
df['year'] = df['Accident_Date'].dt.year
total = pd.pivot_table(data = df, index = 'year', values = ['Number_of_Casualties','Number_of_Vehicles'], aggfunc='sum')
print(total)

# double bar graph
total = total.reset_index()
total.set_index('year')[['Number_of_Casualties','Number_of_Vehicles']].plot(kind = 'bar', color =['pink','skyblue'])
plt.title("\nTotal Casualty and Vehicle Count of 2021 and 2022\n")
plt.xlabel("\nYear")
plt.ylabel("Accident Count\n")
plt.xticks(rotation = 0)
plt.show()
```

### 12. Which dates and months have the most and least number of accident counts?

```python
acc_date = df.groupby('Accident_Date')
dat = acc_date['Accident_Index'].agg(np.size).sort_values().reset_index()
print(dat.iloc[0:3])
print(dat.iloc[-3:])

df['month'] = df['Accident_Date'].dt.month
mon = pd.pivot_table(data = df, index = 'month', values = ['Accident_Index'], aggfunc='count')
print(mon)

# bar plot
mon.plot(kind = 'bar', color = 'cyan', legend = False)
plt.title("\nAccident Count in each Month\n")
plt.xlabel("\nMonths")
plt.ylabel("Accident Count\n")
plt.xticks(rotation = 0)
plt.show()
```

### 18. What is the relationship between the type of vehicle and accident severity? 

```python
veh_sev = df.groupby('Vehicle_Type')['Accident_Severity'].value_counts()
print(veh_sev)

# stacked bar plot
veh_sev.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('\nAccident Severity Distribution by Vehicle Type\n')
plt.xlabel('\nVehicle Type')
plt.ylabel('Accident Count\n')
plt.show()
```

### 19. Categorize the yearly accident count in Urban and Rural areas.

```python
cat = df.groupby('year')['Urban_or_Rural'].value_counts()
print(cat)

# stacked bar graph
cat.plot(kind ='bar', stacked = True)
plt.title("\nYearly Accident Count in Urban and Rural Areas\n")
plt.xlabel("\nYear with Area")
plt.ylabel("Accident Count\n")
plt.show()
```

### 20. What is the vehicle count with speed limit > 30?

```python
vehicle_count = df[df['Speed_limit'] > 30].groupby('Vehicle_Type').size()
print(vehicle_count)

# 
vehicle_count.plot(kind = 'bar', color = 'magenta')
plt.title("\nVehicle Count with speed limit > 30\n")
plt.xlabel("\nVehicle type")
plt.ylabel("Vehicle Count\n")
plt.xticks(rotation = 90)
plt.show()
```

# Summary
#### The analysis provided detailed responses to key questions about road accidents:
##### Accident Trends: Accidents are most frequent during peak traffic hours, with a notable increase during weekends and adverse weather conditions. Urban areas and intersections are identified as high-risk zones.
##### Factors Contributing to Accidents: Speeding, poor visibility (fog or rain), and driver negligence are the leading contributors. Seasonal variations, such as icy roads in winter, also play a role.
##### Severity Analysis: Severe accidents are more common on highways and during nighttime, influenced by higher speeds and lower traffic control measures.
##### High-Risk Locations: Accident hotspots were mapped using spatial analysis, highlighting areas that require immediate safety interventions.
##### Predictive Insights: If predictive modeling was included, the results show promising accuracy in estimating accident severity based on factors like weather, time, and location.
